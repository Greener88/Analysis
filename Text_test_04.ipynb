```python
import csv
import pandas as pd


from konlpy.tag import Mecab
mecab = Mecab()

f = open('220311_Second Opinion_Snall.txt','r',encoding='utf-8')
lines = f.readlines()

line = []
for i in range(len(lines)) : 
    line.append(lines[i])
f.close()



# 특수문자 지우기

import re
compile = re.compile("[^ ㄱ-ㅣ가-힣]+")
for i in range(len(line)):

    a = compile.sub("",line[i])
    line[i] = a
print(line)



# 불용어 제거

stop_word = ["허 참 세상일도", "내면서", "것이다", "요청", "민원", "LH", "민원내용", "안녕하세요"] #사용하지 않을 단어 또는 문장 추가
line = [i for i in line if i not in stop_word]



# 토큰화 하기

result =[]
result = [mecab.nouns(i) for i in line] #명사만 추출
final_result= [r for i in result for r in i]
print(final_result)


# 단어 수 세기

import pandas as pd

korean = pd.Series(final_result).value_counts().head(10)
print("한국어 top 10")
print(korean)
```
